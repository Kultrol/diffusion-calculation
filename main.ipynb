```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Given parameters
n = 3e16  # cm^-3
vTh = 23000  # cm/s
sigmaTot = 1000  # Angstrom^2
mu = 720.6  # g/mol
```

```python
# Convert sigmaTot from Angstrom^2 to cm^2
sigmaTot = sigmaTot * 1e-16  # cm^2
```

```python
# Calculate mean time between collisions (tau)
tau = 1e16 / (n * sigmaTot * vTh)

# Time series from 0 to 60 seconds
time_series = np.linspace(0, 60, 500)  # seconds

# Define a range of temperatures from 100K to 1000K in 100K increments
temperatures = np.arange(100, 1100, 100)
```

```python
# Plot distance traveled over time for each temperature
plt.figure()


plt.figure(figsize=(10, 7), dpi=150)
for Ttrans in temperatures:
    # Calculate diffusion coefficient (D) for each temperature
    D = (41572299.3 * Ttrans * tau) / mu  # cm^2/s
    # Calculate distance traveled over time for each temperature
    distance_traveled = np.sqrt(6 * D * time_series)  # cm
    # Plot the distance traveled
    plt.plot(time_series, distance_traveled, label=f'{Ttrans}K')



plt.style.use('seaborn-v0_8-white')
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Distance (cm)')
plt.title('Distance Traveled vs Time for Varying Temperatures')
plt.legend(title='Temperature (K)', loc='upper left')
plt.grid(True)
plt.show()
```
